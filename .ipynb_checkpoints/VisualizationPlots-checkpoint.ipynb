```python
accDict = {}
for ix in range(4) :
    for name in allF[ix] :
        if '.csv' in name :
            folder = ''
            if 'ER' in name :
                folder = 'mc_chain_ER'
            elif 'BA' in name : 
                folder = 'mc_chain_BA'
            elif 'WS' in name :
                folder = 'mc_chain_WS'
            else : 
                folder = 'mc_chain_MC'
                
            modelName= name[4:].replace("-tag-Test_Accuracy.csv",'')
            df = pd.read_csv('./scalars/'+folder+'/'+name)
#             print name
#             print df.head()
            accDict[modelName] = df['Value'].max()
```

```python
import matplotlib.pyplot as plt
```

```python
def plotter(xfun,Yaxis=aM,Xaxis=accDict,text=''):
    aM = Yaxis
    accDict = Xaxis
    
    mcx = []
    mcy = []
    
    erx = []
    ery = []
    
    bax = []
    bay = []
    
    wsx = []
    wsy = []
    
    maxX = 0
    maxY = 0
    maxName = ''
    for k,v in accDict.iteritems() :
    
        ff = xfun(aM[k])
        color = 'red'
        
        if maxY < v : 
            maxX = ff
            maxY = v
            maxName = k
        
        if 'BA' in k :
            bax.append(ff)
            bay.append(v)
        elif 'ER' in k :
            erx.append(ff)
            ery.append(v)
        elif 'WS' in k:
            wsx.append(ff)
            wsy.append(v)
        else : 
            mcx.append(ff)
            mcy.append(v)
            
#     print maxY
    plt.figure(figsize=(15,6))
    

    plt.scatter(mcx,mcy,c='red',label='MC')
    plt.scatter(erx,ery,c='green',label='ER')
    plt.scatter(bax,bay,c='blue',label='BA')
    plt.scatter(wsx,wsy,c='orange',label='WS')
    plt.ylabel("Test Accuracy")
    plt.annotate("Max Accuracy", xy=(maxX, maxY), xytext=(maxX, maxY-6), arrowprops=dict(arrowstyle="->"))


    
    plt.title(text)
    plt.legend(loc='lower left')
    
    plt.plot()
    
    return maxName
```

```python
def centrality(x):
    centrality = nx.degree_centrality(x)
    ff =  sorted('{:0.2f}'.format(c) for v, c in centrality.items())[-1]
    return ff

def diameterX(x):
    return len(nx.algorithms.dag.dag_longest_path(x))

def densityX(x):
    return nx.density(x)

def maxDegree(x):
    return sorted(v for k,v in x.degree)[-1]

def meanDegree(x) :
    return np.mean([v for k,v in x.degree])

def varDegree(x) :
    return np.var([v for k,v in x.degree])

def medianDegree(x) :
    return np.median([v for k,v in x.degree])

def numNodes(x):
    return len(x.nodes)

def passfn(x):
    return 1
```

```python
def plotHist(fun,x,y,text='',kde=True):
    plt.figure(figsize=(15,6))
    xt = []
    for k,v in y.iteritems() :
        ff = fun(x[k])
        xt.append(ff)
    
    sns.distplot(xt, hist=True, kde=kde, 
                 bins=int(180/5), color = 'darkblue', 
                 hist_kws={'edgecolor':'black'},
                 kde_kws={'linewidth': 4})
    

    plt.title(text)
    plt.plot()
```

```python
plotHist(densityX,aM,accDict,'Density Hist')
```

```python
plotter(densityX,Yaxis=aM,Xaxis=accDict,text='Density')
```

```python
plotHist(centrality,aM,accDict,'Degree Centrality Hist',kde=False)
```

```python
plotter(centrality,aM,accDict,'Degree Centrality')
```

```python
plotHist(diameterX,aM,accDict,'Diameter Hist')
```

```python
plotter(diameterX,aM,accDict,'Diameter')
```

```python
plotHist(maxDegree,aM,accDict,'Max Degree Hist')
```

```python
plotter(maxDegree,aM,accDict,'Max Degree')
```

```python
plotHist(meanDegree,aM,accDict,'Mean Degree Hist')
```

```python
plotter(meanDegree,aM,accDict,'Mean Degree')
```

```python
plotHist(varDegree,aM,accDict,'Variance Degree Hist')
```

```python
plotter(varDegree,aM,accDict,'Variance Degree')
```

```python
plotter(medianDegree,aM,accDict,'Median Degree')
```

```python
plotter(numNodes,aM,accDict,'Total Number of Nodes')
```
